```python
import numpy as np
def read_hr(fname, thresh = 0.001):
    f = open(fname,"r")
    f.readline()
    num_wann = int(f.readline())
    nR = int(f.readline())
    ndeg = []
    while len(ndeg) < nR:
        ndeg += [ int(i) for i in f.readline().split() ]
    ndeg = np.array(ndeg)
    HR = {}
    for ir in range(nR):
        for i in range(num_wann):
            for j in range(num_wann):
                l = f.readline().split()
                a = float(l[-2]) + 1j *float(l[-1])
                if abs(a)>thresh:
                    i = tuple(int(_) for _ in l[:5])
                    HR[i] = a
    return HR

H_triv=read_hr('hald_hr_triv.dat', 0.01)
H_top = read_hr('hald_hr_top.dat', 0.01)

for H in H_triv, H_top:
    lst = [ [i,h,abs(h)] for i, h in H.items() ]
    srt = np.argsort( [l[-1] for l in lst])[-1::-1]
    print (len(srt))
    for i in srt:
        print (lst[i])
    print()


    


```

```python
from wannierberri import Path, Grid, evaluate_k_path, run
from wannierberri.system import System_PythTB
from wannierberri.models import Haldane_ptb
from wannierberri import calculators


model_haldane_triv = Haldane_ptb (delta = -0.047, hop1 = -0.33, hop2 = 0.02, phi=0)
model_haldane_topo = Haldane_ptb (delta = -0.055, hop1 = -0.54, hop2 = 0.056, phi=np.pi/3)

model_haldane_triv = Haldane_ptb (delta = -0.05, hop1 = -0.4, hop2 = 0.06, phi=0)
model_haldane_topo = Haldane_ptb (delta = -0.05, hop1 = -0.4, hop2 = 0.08, phi=np.pi/3)


system_haldane_topo = System_PythTB(model_haldane_topo)
system_haldane_triv = System_PythTB(model_haldane_triv)



path = Path(system = system_haldane_triv, nodes = [[0,0,0], [2/3,1/3,0],[1/2,0,0], [0,0,0],
                                                   [-2/3,-1/3,0], [-1/2,0,0], [0,0,0]]
                                                
            , labels = ['$\Gamma$', '$K$', '$M$', '$\Gamma$', '$K\'$', '$M\'$', '$\Gamma$'],
                   length = 300)

bands_triv = evaluate_k_path(system_haldane_triv, path=path)
bands_topo = evaluate_k_path(system_haldane_topo, path=path)
import matplotlib.pyplot as plt
bands_triv.plot_path_fat(path=path, label='trivial' ,close_fig=False, show_fig=False , Emin=-2, Emax=2, linecolor='blue')
bands_topo.plot_path_fat(path=path, label='topological',close_fig=False, show_fig=False, Emin=-2, Emax=2, linecolor='red')
plt.show()

```

```python

efermi = np.linspace(-2, 2, 1000)
omega = np.linspace(0, 2, 1000)
calc = {
    "dos" : calculators.static.DOS(Efermi=efermi, tetra=True),
    "ahc" : calculators.static.AHC(Efermi=efermi, tetra=True),
    "shift_current" : calculators.dynamic.ShiftCurrent(Efermi=np.array([0]), omega = omega, sc_eta=0.01),
    "tabulate" : calculators.tabulate.TabulatorAll({}, mode="grid")
}



grid = Grid(system_haldane_triv, length=100, NKFFT =[10,10,1])
res_triv = run(system_haldane_triv, grid, calculators=calc, print_Kpoints=True)
res_topo = run(system_haldane_topo, grid, calculators=calc)

```

```python


plt.plot(efermi, res_triv.results['dos'].data, label='trivial')
plt.plot(efermi, res_topo.results['dos'].data, label='topological')
plt.title('DOS')
plt.ylim([0 , res_topo.results['dos'].data.max()*1.1])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS (states/eV)')
plt.legend()
plt.show()

```

```python

ahc_quant =  7.74809172e-5
data = res_triv.results['ahc'].data[:,2]* system_haldane_triv.real_lattice[2,2] * 1e-10 / ahc_quant
plt.plot(efermi, data)
data = res_topo.results['ahc'].data[:,2]* system_haldane_triv.real_lattice[2,2] * 1e-10 / ahc_quant
plt.plot(efermi, data)
plt.title('AHC')
plt.xlabel('Energy (eV)')
plt.ylabel('AHC (2$e^2$/h)')
plt.legend(['trivial', 'topological'])
plt.ylim([0,0.7])
plt.show()
```

```python

data = res_triv.results['shift_current'].data[0,:,1,1,1] #* system_haldane_triv.real_lattice[2,2] * 1e-10
plt.plot(omega, data, label='trivial yy')
data = res_triv.results['shift_current'].data[0,:,0,0,0] #* system_haldane_triv.real_lattice[2,2] * 1e-10
plt.plot(omega, data, label='trivial xx')

data = res_topo.results['shift_current'].data[0,:,1,1,1] #* system_haldane_topo.real_lattice[2,2] * 1e-10
plt.plot(omega, data, label='topo yy')
data = res_topo.results['shift_current'].data[0,:,0,0,0] #* system_haldane_topo.real_lattice[2,2] * 1e-10
plt.plot(omega, data, label='topo xx')
plt.xlabel('Frequency (eV)')
plt.ylabel('Shift current ')
plt.legend()
plt.title('Shift current')

plt.show()
```
